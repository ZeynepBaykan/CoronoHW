# Corona paper analysis

## "ab.csv" abstract part of include corona related papers. Each line is taken from different paper.

```python

#Download data
import numpy as np
import pandas as pd
import nltk
from nltk.corpus import stopwords
from collections import Counter

x = pd.read_csv('ab.csv', delimiter ='##', encoding='utf-8')  
#x
y= x['Content']

arr = []
for i in y:
    arr.append(i)

stop=stopwords.words('english')
df = pd.DataFrame(y)
df
```

```python
df['Content']=df['Content'].apply(lambda x:" ".join(x.lower() for x in x.split()))
df['Content']=df['Content'].apply(lambda x: " ".join(x for x in x.split()if x not in stop))
```

# What is the most frequent words?

```python
Counter(" ".join(df['Content']).split()).most_common(10)
```

```python
sentences_list = []
word = "treatment"

new_arr2= []

for i in df['Content']:
    sentences_list = i.split(".")
    for k in sentences_list:
        words = k.split()
        if word in words:
            new_arr2.append(k)
            #count2 = Counter(k.split())
            #most_freq = count2.most_common(1)
            #print(most_freq)
            #print(k)
            
df3 = pd.DataFrame(new_arr2)
df3


Counter(" ".join(df3[0]).split()).most_common(10) #most common 10 words from the sentences that include treatment word.


        
```

```python
# Show frequency of most common words  

ax= Counter(" ".join(df['Content']).split()).most_common(5)

ma=[]

numb=[]

for i in range(len(ax)):
    ma.append(ax[i][0])

for i in range(len(ax)):
    numb.append(ax[i][1])
    
import matplotlib.pyplot as plt



def plot_bar_x():
    # this is for plotting purpose
    index = np.arange(len(ma))
    plt.bar(index, numb)
    plt.xlabel('Most Common Words', fontsize=10)
    plt.ylabel('Frequency', fontsize=10)
    plt.xticks(index, ma, fontsize=10, rotation=30)
    plt.title('Frequency of Most Common Words')
    plt.show()

plot_bar_x()

```
